```python
#連想配列を作成
personal_data = {'name': '侍太郎', 'age': 30, 'gender': '男性'}
print(personal_data)
personal_data['name'] = '足立篤治'
personal_data['age'] = 52
print(personal_data)
personal_data['address'] = '神奈川県'
print(personal_data)
personal_data.pop('age')
print(personal_data)
```
